```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.load('Plot_data/X.npy')
Y=np.load('Plot_data/Y.npy')
chi2_training_noR=np.load('Plot_data/chi2_training_noR.npy') 
chi2_validation_noR=np.load('Plot_data/chi2_validation_noR.npy') 
chi2_training_R=np.load('Plot_data/chi2_training_R.npy',allow_pickle=True) 
chi2_validation_R=np.load('Plot_data/chi2_validation_R.npy',allow_pickle=True) 
```

```python
frame=(16,8)
```

# Figure 3

# a

```python
plt.figure(figsize=frame)
levels = np.arange(0, 12, 1)
plt.contourf(X,Y,chi2_training_noR, levels=levels, cmap='viridis')
cbar = plt.colorbar(format = '%1.0f')
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
cbar.set_label(r'$\chi^2$',rotation=0, labelpad=20,weight='bold',fontsize=18)
plt.contour(X,Y,chi2_training_noR, levels=levels,colors='black',linewidths=0.5)
levels = np.arange(0, 1, 0.1)
plt.contour(X,Y,chi2_training_noR, colors='black', levels=levels,linewidths=0.25)
\
plt.xlabel(r'$\tilde{\alpha}$',fontsize=22)
plt.ylabel(r'$\tilde{\beta}$',fontsize=22)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
plt.show() 
```

# b

```python
plt.figure(figsize=frame)
levels = np.arange(0, 12, 1)
plt.contourf(X,Y,chi2_validation_noR,levels=levels, cmap='viridis')
cbar = plt.colorbar(format = '%1.0f')
plt.contour(X,Y,chi2_validation_noR, levels=levels, colors='black',linewidths=0.5)
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
cbar.set_label(r'$\chi^2$',rotation=0, labelpad=20,weight='bold',fontsize=18)
levels = np.arange(0, 1, 0.1)
plt.contour(X,Y,chi2_validation_noR, levels=levels, colors='black',linewidths=0.5)
levels = np.arange(0, 1, 0.05)
plt.contour(X,Y,chi2_validation_noR, colors='black', levels=levels,linewidths=0.5)

row,col=np.where(chi2_validation_noR == np.min(chi2_validation_noR))
plt.plot(X[col[-1]], Y[row[-1]], color='red',marker='*',markersize=25,markeredgecolor='mistyrose')

                    
plt.xlabel(r'$\tilde{\alpha}$',fontsize=22)
plt.ylabel(r'$\tilde{\beta}$',fontsize=22)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
plt.show() 

print('alpha: ',X[col[0]],'beta: ', Y[row[0]])
```

# Figure 4

# a

```python
from matplotlib.ticker import FuncFormatter, MultipleLocator


experimental_max_min=np.load('SI_saves/experimental_max_min.npy')

beta_min_coupl=experimental_max_min[0]
beta_max_coupl=experimental_max_min[1]

params1=np.load('Plot_data/beta_ThisStudy.npy')
params2=np.load('Plot_data/beta_ThisStudyNoR.npy')
params3=np.load('Plot_data/beta_Lankhorst.npy')
params4=np.load('Plot_data/beta_Mooren.npy')
params5=np.load('Plot_data/beta_Lee.npy')
params6=np.load('Plot_data/beta_Blackburn.npy')
params7=np.load('Plot_data/beta_Altona.npy')
ave_lit_curve=np.load('Plot_data/beta_ave_lit_curve.npy')


cmap = plt.cm.get_cmap('cool')
ncomparisons=5
marker_list=['s','^','P','D','*','H','X','v']

x=np.linspace(-np.pi,np.pi)

fig, ax = plt.subplots(figsize=frame)

plt.axhspan(beta_min_coupl, beta_max_coupl, color='darkgray', alpha=0.5, lw=0) #, label='Exp. range')

plt.plot(x,params3,linewidth=2,color=cmap((0+1)/ncomparisons),marker=marker_list[0],label='Lankhorst (1984)')
plt.plot(x,params4,linewidth=2,color=cmap((1+1)/ncomparisons),marker=marker_list[1],label='Mooren (1994)')
plt.plot(x,params5,linewidth=2,color=cmap((2+1)/ncomparisons),marker=marker_list[2],label='Lee (1976)')
plt.plot(x,params6,linewidth=2,color=cmap((3+1)/ncomparisons),marker=marker_list[3],label='Blackburn (1973)')
plt.plot(x,params7,linewidth=2,color=cmap((4+1)/ncomparisons),marker=marker_list[4],label='Altona (1982)')
plt.plot(x,params2,'--',color='dimgrey',linewidth=2,label=r'This Study $\chi_{red}^2$')
plt.plot(x,params1,'o-',color='black',linewidth=2,label=r'This Study $\chi_{red}^2+\frac{\tilde{\gamma}R}{3}$')


ave_array=np.mean(ave_lit_curve,axis=0)
std_array=np.std(ave_lit_curve,axis=0)
max_lit=np.max(ave_lit_curve,axis=0)
min_lit=np.min(ave_lit_curve,axis=0)
plt.fill_between(x, min_lit-std_array, max_lit+std_array,color='blue',alpha=0.2)
plt.ylim(-3,39.5)  

plt.xlabel(r'HCOP ($\beta \pm \frac{2\pi}{3}$)',fontsize=22)
plt.ylabel('J-coupling in Hz',fontsize=22)
plt.legend(ncol=2,fontsize=22)
plt.yticks(fontsize=22)

ax.tick_params(axis='both', which='both', labelsize=22)

ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi))
ax.xaxis.set_minor_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_minor_locator(MultipleLocator(base=np.pi/2))
        
plt.tight_layout()
plt.show()
```

# Figure 5

# a

```python
resolution=30
X= np.geomspace(1e1, 1e10, num=resolution)

fig,ax = plt.subplots(figsize=frame)

cmap = plt.cm.get_cmap('cool')
c=cmap(1/5) 

orig_error=np.load('Plot_data/train_jcoupl_orig_error.npy')
new_error=np.load('Plot_data/train_jcoupl_new_error.npy')
alphas=np.load('Plot_data/alphas.npy')
    
    
plt.plot(X,new_error,'-o',color=c,linewidth=5,markersize=15,zorder=15,label='opt. J-coupling')
marker_loc=np.where(X==alphas[11])[0][0]
plt.plot(X[marker_loc],new_error[marker_loc], color=c,marker='*',alpha=1,markersize=25,markeredgecolor='black',zorder=20)
plt.plot(X[marker_loc],orig_error[marker_loc], color='black',marker='*',alpha=1,markersize=25,markeredgecolor='black',zorder=20)

plt.plot(X,orig_error,'--',color='black',linewidth=5,markersize=7,label='orig. J-coupling')

print_orig_err=orig_error[-1]
print('error in orig ensemble: ',print_orig_err)
print_orig_err=orig_error[4]
print('error with orig. Karplus parameters: ',print_orig_err)
print_opt_err=new_error[4]
print('error with opt. Karplus parameters: ',print_opt_err)


plt.ylabel(r'$\chi_{red}^2$',fontsize=22)
plt.xlabel(r'$\alpha$',fontsize=22)
plt.xscale('log')
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.legend(fontsize=22)

plt.tight_layout()
plt.show()
```
